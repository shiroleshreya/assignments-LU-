```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#Question 1:

#plot a line plot between a and b:

a=np.arange(40,50)

b=np.arange(50,60)

#In [3]:
a=np.arange(40,50)
b=np.arange(50,60)

plt.xlabel("X-Axis", color='red')
plt.ylabel("Y-Axis", color='Red')
plt.plot(a,b,'b*-',linewidth=1,markersize=8)
plt.show()
```

```python
#Question 2: Plot a line plot showing the sales trend in company 1 and 2:

days = [1,2,3,4,5,6,7] #days of d week sales_1 = [160,150,140,145,175,165,180] #sales of company1 sales_2 = [70,90,160,150,140,145,175] #sales of company2

#In [4]:
days = [1,2,3,4,5,6,7]
sales_1 = [160,150,140,145,175,165,180] #company1
sales_2 = [70,90,160,150,140,145,175] #company2

plt.xlabel("X-Axis", color='blue')
plt.ylabel("Y-Axis", color='blue')

plt.plot(days, sales_1, 'bo-', markersize=12, linewidth=2)
plt.plot(days, sales_2, 'ro-', markersize=12, linewidth=2)
plt.show()
```
